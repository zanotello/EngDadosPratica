```python
import duckdb   
```

```python
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
df = con.execute("SELECT * FROM produtos").fetchdf()
df.head(10)
```

```python
df2 = df.drop(columns=['fornecedor', 'categoria'])
df2.head(10)
```

```python
con.execute("""CREATE TABLE IF NOT EXISTS dim_produtos (
        id BIGINT,
        nome VARCHAR,
        preco FLOAT
    )""")
```

```python
con.execute("INSERT INTO dim_produtos SELECT * FROM df2")
df = con.execute("SELECT * FROM dim_produtos").fetchdf()
df.head(10)
```

```python
con.close()
```
